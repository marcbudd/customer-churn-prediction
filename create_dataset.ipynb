# Merge dataframes

```python
df = df_churn[['Customer ID', 'Churn']]
df = df.rename(columns={'CustomerID': 'Customer ID'})

# Merge the dataframes

# Demographics
doubled_columns = ['Count']
temp = df_demographics.drop(columns=doubled_columns)
df = df.merge(temp, on='Customer ID', how='inner')

# Location
doubled_columns = ['Count']
temp = df_location.drop(columns=doubled_columns)
df = df.merge(temp, on='Customer ID', how='inner')

# Population
temp = df_population.drop(columns=['ID'])
temp = temp.rename(columns={'Population': 'Zip Code Population'})
df = df.merge(temp, on='Zip Code', how='inner')

# Services
temp = df_services.drop(columns=['Service ID'])
df = df.merge(temp, on='Customer ID', how='inner')

# Status
needed_columns = ['Customer ID', 'Satisfaction Score']
temp = df_status[needed_columns]
df = df.merge(temp, on='Customer ID', how='inner')

df.columns
```

```python
df.to_csv('data/churn_data.csv', index=False)
```
